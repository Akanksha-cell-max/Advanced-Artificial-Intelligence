```python
import numpy as np
import matplotlib.pyplot as plt

class SimpleRobot:
    def __init__(self, x=0, y=0, heading=0):
        # Position and orientation (in degrees)
        self.x = x
        self.y = y
        self.heading = heading  # 0 = facing right

        self.path = [(x, y)]  # store path history

    def move_forward(self, distance):
        # Move in the current heading direction
        rad = np.radians(self.heading)
        self.x += distance * np.cos(rad)
        self.y += distance * np.sin(rad)
        self.path.append((self.x, self.y))

    def turn_left(self, angle):
        self.heading = (self.heading + angle) % 360

    def turn_right(self, angle):
        self.heading = (self.heading - angle) % 360

    def stop(self):
        print(f"Robot stopped at ({self.x:.2f}, {self.y:.2f}) facing {self.heading:.2f}°.")

    def plot_path(self):
        path = np.array(self.path)
        plt.plot(path[:, 0], path[:, 1], marker='o')
        plt.quiver(path[:-1, 0], path[:-1, 1],
                   path[1:, 0] - path[:-1, 0],
                   path[1:, 1] - path[:-1, 1],
                   scale_units='xy', angles='xy', scale=1, color='r')
        plt.title("Robot Path Visualization")
        plt.xlabel("X Position")
        plt.ylabel("Y Position")
        plt.grid()
        plt.axis('equal')
        plt.show()

# Create a robot
robot = SimpleRobot()

# Control Program: sequence of moves
robot.move_forward(5)
robot.turn_left(90)
robot.move_forward(3)
robot.turn_right(45)
robot.move_forward(4)
robot.turn_right(45)
robot.move_forward(2)
robot.stop()

# Visualize the movement path
robot.plot_path()
```
